```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
import re
```

```python
#Cargar el csv de la data limpia
data = pd.read_csv('../data/processed/Airbnb_Data_Processed_2.csv')
data.head()
```

```python
data.columns
```

```python
#ordinal encoder sklearn para property_type, room_type, bed_type, cancellation_policy
#Se cambian las columnas actuales con los resultados del ordinal encoder
encoder1 = OrdinalEncoder()
encoder2 = OrdinalEncoder()
encoder3 = OrdinalEncoder()
encoder4 = OrdinalEncoder()
data['property_type'] = encoder1.fit_transform(data[['property_type']]).astype(int)
data['room_type'] = encoder2.fit_transform(data[['room_type']]).astype(int)
data['bed_type'] = encoder3.fit_transform(data[['bed_type']]).astype(int)
data['cancellation_policy'] = encoder4.fit_transform(data[['cancellation_policy']]).astype(int)
```

```python
data.head()
```

```python
#Verificar consistencia de tipos de datos viendo los valores unicos de cada columna
print(data.dtypes)
```

```python
for column in data.columns:
    if column != 'log_price' and column != 'amenities':   
        print(column)
        print(data[column].unique())
```

```python
#Ver los valores únicos en cada columna por medio de un set de amenities
#La data en ameninities esta de la forma {a,b,c}
amenities = set()
for i in data['amenities']:
    # print(i)
    for j in i[1:-1].split(','):
        # print(j)
        # Expresión regular para quitar espacios y comillas al inicio y final
        j_sin_comillas = re.sub(r'^\s*"?(.*?)"?\s*$', r'\1', j)

        amenities.add(j_sin_comillas)
print(sorted(amenities))
```

```python
#Eliminar todos los que empiecen por 'translation missing'
amenities = {i for i in amenities if not i.startswith('translation missing')}
#Al amenities vacío se le cambia por 'No amenities'
amenities.discard('')
amenities.add('No amenities')
amenities=sorted(list(amenities))
print(amenities)
print("Total amenities: ", len(amenities))
```

```python

#Genera una columna por cada amenity, si esta en la lista de amenities, se pone 1, sino 0
for i in amenities:
    data[i] = data['amenities'].apply(lambda x: 1 if i in x else 0)

```

```python
#Tratamiento de nombres de cada columna poe medio de
#Expresion regular que quita cualquier espacio intermedio y guion (-) y lo reemplaza por un guion bajo
#Además pone todo en minúsculas
data.columns = [re.sub(r'\s+', '_', i.lower()) for i in data.columns]
print(data.columns) 
data.head()
```

```python
#Eliminar la columna amenities
data = data.drop(columns=['amenities'])
#Guardar la data en un csv
data.to_csv('../data/processed/Airbnb_Data_Processed_3_Models.csv', index=False)
```
